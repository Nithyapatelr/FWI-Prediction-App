```python
# Run this in Google Colab (no extra installs required)
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from scipy.stats import pointbiserialr, chi2_contingency
import math
import matplotlib.pyplot as plt
import seaborn as sns

# --- ASSUMPTION: df is your DataFrame and 'Churn' is 0/1 ---
# If not loaded, load your dataframe first (example):
# df = pd.read_csv('/path/to/your.csv')

# 1) Prepare X and y
X = df.drop(columns=['Churn'], errors='ignore').copy()
y = df['Churn'].copy()

# Keep copy of original columns for plotting
orig_cols = X.columns.tolist()

# 2) Create a numeric copy for mutual info & RF
X_num = X.copy()

# Label-encode object/categorical columns for MI and RF (simple and safe)
cat_cols = X_num.select_dtypes(include=['object', 'category']).columns.tolist()
for col in cat_cols:
    X_num[col] = LabelEncoder().fit_transform(X_num[col].astype(str))

# 3) Mutual Information (non-linear, works for mixed data after label-encoding)
mi = mutual_info_classif(X_num, y, discrete_features='auto', random_state=42)
mi_ser = pd.Series(mi, index=X_num.columns).sort_values(ascending=False)

# 4) Random Forest importance (model-based)
rf = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced', n_jobs=-1)
rf.fit(X_num, y)
rf_imp = pd.Series(rf.feature_importances_, index=X_num.columns).sort_values(ascending=False)

# 5) Point-biserial for numeric features (Churn is binary)
num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
pb_scores = {}
for col in num_cols:
    try:
        corr, pval = pointbiserialr(y, X[col])
        pb_scores[col] = corr
    except Exception:
        pb_scores[col] = np.nan
pb_ser = pd.Series(pb_scores).sort_values(key=lambda s: s.abs(), ascending=False)

# 6) Cramer's V for categorical features vs Churn
def cramers_v(x, y):
    confusion = pd.crosstab(x, y)
    if confusion.size == 0:
        return 0.0
    chi2 = chi2_contingency(confusion)[0]
    n = confusion.sum().sum()
    if n == 0:
        return 0.0
    phi2 = chi2 / n
    r, k = confusion.shape
    # bias correction
    phi2corr = max(0, phi2 - ((k-1)*(r-1)) / (n-1))
    rcorr = r - ((r-1)**2) / (n-1)
    kcorr = k - ((k-1)**2) / (n-1)
    denom = min((kcorr-1), (rcorr-1))
    if denom <= 0:
        return 0.0
    return math.sqrt(phi2corr / denom)

cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
cramer_scores = {}
for col in cat_cols:
    try:
        cramer_scores[col] = cramers_v(X[col].astype(str), y)
    except Exception:
        cramer_scores[col] = 0.0
cramer_ser = pd.Series(cramer_scores).sort_values(ascending=False)

# 7) Build combined DataFrame of scores (aligning all feature names)
all_features = sorted(set(X_num.columns.tolist()))
summary = pd.DataFrame(index=all_features)
summary['mutual_info'] = mi_ser.reindex(all_features).fillna(0)
summary['rf_importance'] = rf_imp.reindex(all_features).fillna(0)
summary['point_biserial'] = pb_ser.reindex(all_features).fillna(0)
# Cramer's V is only meaningful for original categorical names;
# For label-encoded categorical features, their name is same, so reindex:
summary['cramers_v'] = cramer_ser.reindex(all_features).fillna(0)

# 8) Create a simple aggregated rank (you can weight metrics as desired)
# Normalize columns to 0-1 then average
def norm_series(s):
    if s.max() == s.min():
        return s * 0
    return (s - s.min()) / (s.max() - s.min())

summary_norm = summary.copy()
for c in summary.columns:
    summary_norm[c] = norm_series(summary[c].abs())  # use absolute for correlations
summary_norm['aggregate_score'] = summary_norm.mean(axis=1)

summary = summary.join(summary_norm['aggregate_score'])
summary = summary.sort_values(by='aggregate_score', ascending=False)

# 9) Print top features
print("Top features by aggregate predictive score:\n")
display(summary.head(20))

# 10) Visualize top 10 features (aggregate)
topk = summary.head(10)
plt.figure(figsize=(8,6))
sns.barplot(x=topk['aggregate_score'].values, y=topk.index, palette='viridis')
plt.title('Top features by aggregated predictive score')
plt.xlabel('Aggregate (normalized) score')
plt.show()

# 11) Also show RF and MI side-by-side for top features
top_feats = topk.index.tolist()
fig, axes = plt.subplots(1,2, figsize=(14,6))
sns.barplot(x=rf_imp.reindex(top_feats).values, y=top_feats, ax=axes[0])
axes[0].set_title('RandomForest importance (top features)')
sns.barplot(x=mi_ser.reindex(top_feats).values, y=top_feats, ax=axes[1])
axes[1].set_title('Mutual Information (top features)')
plt.tight_layout()
plt.show()
```

```python
# Feature engineering: LoyaltyScore, AddOnCount / NoAddOns, TechEngagement, PriceVsDuration
import numpy as np
import pandas as pd

# --- 0. Helper: safe column check
def ensure_cols_exist(df, cols):
    return [c for c in cols if c in df.columns]

# --- 1. Create/ensure ContractDuration (in months)
# If 'Contract' is textual, map values. If numeric 1/2/3, map to months (1,12,24).
if 'ContractDuration' not in df.columns:
    if 'Contract' in df.columns:
        if df['Contract'].dtype == object or df['Contract'].dtype.name == 'category':
            # map textual contract labels (common variants)
            mapping = {
                'Month-to-month': 1, 'Month to month': 1, 'month-to-month': 1, 'month to month': 1,
                'One year': 12, 'One-year': 12, 'one year': 12, '1 Year': 12,
                'Two year': 24, 'Two-year': 24, 'two year': 24, '2 Year': 24
            }
            df['ContractDuration'] = df['Contract'].map(mapping).astype(float)
        else:
            # numeric contract column: assume coded 1,2,3 -> map to months if values fit
            unique_vals = sorted(df['Contract'].dropna().unique())
            if set(unique_vals).issubset({0,1,2,3,4}) or max(unique_vals) <= 3:
                # assume 1 -> month-to-month, 2 -> one year, 3 -> two year
                map_numeric = {0:0, 1:1, 2:12, 3:24}
                df['ContractDuration'] = df['Contract'].map(map_numeric).astype(float)
            else:
                # assume it's already months (e.g., 1,12,24 directly) or custom numeric; copy as-is
                df['ContractDuration'] = pd.to_numeric(df['Contract'], errors='coerce').astype(float)
    else:
        # Try to infer from dummy columns (e.g., Contract_One year, Contract_Two year)
        c_one = 'Contract_One year'
        c_two = 'Contract_Two year'
        c_month = 'Contract_Month-to-month'
        found = False
        if c_one in df.columns or c_two in df.columns or c_month in df.columns:
            df['ContractDuration'] = 1  # default month-to-month
            if c_one in df.columns:
                df.loc[df[c_one] == 1, 'ContractDuration'] = 12
                found = True
            if c_two in df.columns:
                df.loc[df[c_two] == 1, 'ContractDuration'] = 24
                found = True
            if c_month in df.columns:
                df.loc[df[c_month] == 1, 'ContractDuration'] = 1
                found = True
        if not found:
            # fallback: create with NaN so downstream code knows it's missing
            df['ContractDuration'] = np.nan

# If any ContractDuration ended up zero, replace with NaN to avoid divide-by-zero (we'll handle later)
df['ContractDuration'].replace({0: np.nan}, inplace=True)

# --- 2. AddOnCount and NoAddOns
# Add-on columns we expect (modify the list if your dataset has different names)
addon_candidates = ['OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
addon_cols = ensure_cols_exist(df, addon_candidates)

if len(addon_cols) == 0:
    print("Warning: no expected add-on columns found. AddOnCount / NoAddOns not created.")
else:
    # ensure they are numeric (0/1)
    for c in addon_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0).astype(int)
    df['AddOnCount'] = df[addon_cols].sum(axis=1)               # number of active add-ons
    df['NoAddOns'] = (len(addon_cols) - df['AddOnCount']).astype(int)

# --- 3. TechEngagement (count of tech-related services)
tech_cols_candidates = ['OnlineBackup', 'DeviceProtection', 'TechSupport']
tech_cols = ensure_cols_exist(df, tech_cols_candidates)

if len(tech_cols) == 0:
    print("Warning: no expected tech-engagement columns found. TechEngagement not created.")
else:
    df['TechEngagement'] = df[tech_cols].sum(axis=1)

# --- 4. PriceVsDuration (MonthlyCharges divided by ContractDuration)
# Ensure MonthlyCharges numeric
if 'MonthlyCharges' not in df.columns:
    raise KeyError("MonthlyCharges column not found in df.")
else:
    df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')

# Compute PriceVsDuration safely
# If ContractDuration is NaN, set PriceVsDuration to NaN as well
df['PriceVsDuration'] = df['MonthlyCharges'] / df['ContractDuration']
# Optionally cap or replace infinities
df['PriceVsDuration'].replace([np.inf, -np.inf], np.nan, inplace=True)

# --- 5. LoyaltyScore = (ContractDuration * tenure) / 12
# Ensure tenure numeric
if 'tenure' not in df.columns:
    raise KeyError("tenure column not found in df.")
else:
    df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce').fillna(0)

df['LoyaltyScore'] = (df['ContractDuration'] * df['tenure']) / 12.0
# If ContractDuration is NaN, LoyaltyScore will be NaN; replace NaN loyalty with tenure-based fallback if desired:
# df['LoyaltyScore'] = df['LoyaltyScore'].fillna(df['tenure'] / 12.0)

# --- 6. Quick sanity prints
print("\nCreated feature columns (showing head):")
cols_to_show = ['ContractDuration', 'LoyaltyScore', 'AddOnCount', 'NoAddOns', 'TechEngagement', 'PriceVsDuration']
cols_to_show = [c for c in cols_to_show if c in df.columns]
display(df[cols_to_show].head(8))

print("\nSummary statistics:")
display(df[cols_to_show].describe().T)

# Optional: check correlation with Churn immediately
if 'Churn' in df.columns:
    corr = df[cols_to_show + ['Churn']].corr(method='spearman')['Churn'].sort_values(ascending=False)
    print("\nSpearman correlation of new features with Churn:")
    display(corr)
else:
    print("\nChurn column not found; skipping correlation check.")
```

```python
# ================================
# ðŸ”¹ IMPORTS & SETUP
# ================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, classification_report, confusion_matrix

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from catboost import CatBoostClassifier

# ================================
# ðŸ”¹ DATA PREPARATION
# ================================
X = df.drop(columns=['Churn'])
y = df['Churn']

# Optional scaling (Logistic Regression benefits most)
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[X.columns] = scaler.fit_transform(X)

# Train-test split (stratified to preserve churn ratio)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

print("Train:", X_train.shape, " Test:", X_test.shape)
print("Churn distribution (train):\n", y_train.value_counts(normalize=True))

# ================================
# ðŸ”¹ 1. Logistic Regression (Balanced)
# ================================
log_reg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
y_prob_lr = log_reg.predict_proba(X_test)[:,1]

# ================================
# ðŸ”¹ 2. Random Forest
# ================================
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    class_weight='balanced_subsample',
    random_state=42
)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

# ================================
# ðŸ”¹ 3. XGBoost
# ================================
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]
xgb = XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    learning_rate=0.05,
    n_estimators=500,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    random_state=42
)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:,1]

# ================================
# ðŸ”¹ 4. LightGBM / CatBoost
# ================================
# LightGBM
lgbm = lgb.LGBMClassifier(
    learning_rate=0.05,
    n_estimators=500,
    class_weight='balanced',
    random_state=42
)
lgbm.fit(X_train, y_train)
y_pred_lgb = lgbm.predict(X_test)
y_prob_lgb = lgbm.predict_proba(X_test)[:,1]

# CatBoost (silent training)
cat = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    class_weights=[1, 3],
    verbose=0,
    random_state=42
)
cat.fit(X_train, y_train)
y_pred_cat = cat.predict(X_test)
y_prob_cat = cat.predict_proba(X_test)[:,1]

# ================================
# ðŸ”¹ 5. Stacking (XGBoost + Logistic Regression)
# ================================
stack_model = StackingClassifier(
    estimators=[('xgb', xgb), ('rf', rf)],
    final_estimator=LogisticRegression(max_iter=1000, class_weight='balanced'),
    cv=5
)
stack_model.fit(X_train, y_train)
y_pred_stack = stack_model.predict(X_test)
y_prob_stack = stack_model.predict_proba(X_test)[:,1]

# ================================
# ðŸ”¹ MODEL EVALUATION FUNCTION
# ================================
def evaluate_model(y_true, y_pred, y_prob, model_name):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob)
    print(f"\nðŸ“Š {model_name} Evaluation:")
    print(classification_report(y_true, y_pred, digits=3))
    return pd.Series({
        'Model': model_name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-score': f1,
        'ROC-AUC': auc
    })

# Evaluate all models
results = pd.DataFrame([
    evaluate_model(y_test, y_pred_lr, y_prob_lr, "Logistic Regression (Balanced)"),
    evaluate_model(y_test, y_pred_rf, y_prob_rf, "Random Forest"),
    evaluate_model(y_test, y_pred_xgb, y_prob_xgb, "XGBoost"),
    evaluate_model(y_test, y_pred_lgb, y_prob_lgb, "LightGBM"),
    evaluate_model(y_test, y_pred_cat, y_prob_cat, "CatBoost"),
    evaluate_model(y_test, y_pred_stack, y_prob_stack, "Stacking (XGB + LR)")
])

# ================================
# ðŸ”¹ RESULTS COMPARISON
# ================================
plt.figure(figsize=(10,6))
sns.barplot(data=results.melt(id_vars='Model', var_name='Metric', value_name='Score'),
            x='Metric', y='Score', hue='Model', palette='coolwarm')
plt.title("ðŸ“Š Model Comparison on Churn Prediction Metrics", fontsize=14)
plt.xticks(rotation=45)
plt.ylim(0,1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("\nâœ… Final Model Performance Summary:")
display(results.sort_values(by='ROC-AUC', ascending=False))
```

```python
# Full pipeline: tune BRF, tune XGBoost, build stacked ensemble, validate
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from time import time

from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV, cross_val_score, cross_val_predict
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, classification_report, confusion_matrix)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from scipy.stats import randint, uniform

# Models (imblearn + xgboost)
from imblearn.ensemble import BalancedRandomForestClassifier
from xgboost import XGBClassifier

# ---------------------------
# 0) Prepare dataset
# ---------------------------
assert 'Churn' in df.columns, "df must contain 'Churn'"

# features & target
X = df.drop(columns=['Churn'])
y = df['Churn']

# train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

print("Train:", X_train.shape, "Test:", X_test.shape)
print("Churn distribution (train):")
print(y_train.value_counts(normalize=True))

# Use StratifiedKFold for CV
cv_folds = 5
cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)

# ---------------------------
# 1) TUNE BalancedRandomForest (BRF)
# ---------------------------
brf = BalancedRandomForestClassifier(random_state=42, n_jobs=-1)

# Parameter distributions (reasonable ranges)
brf_param_dist = {
    'n_estimators': [200, 300, 400, 600],
    'max_depth': [5, 8, 12, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', 0.5, None],
    # sampling_strategy controls minority sampling - keep defaults mostly
    'sampling_strategy': ['auto', 0.7, 1.0]
}

print("\nTuning BalancedRandomForest (this may take several minutes)...")
rs_brf = RandomizedSearchCV(
    estimator=brf,
    param_distributions=brf_param_dist,
    n_iter=16,                # increase for more thorough search
    scoring='roc_auc',
    cv=cv,
    verbose=1,
    random_state=42,
    n_jobs=-1,
    return_train_score=False
)

t0 = time()
rs_brf.fit(X_train, y_train)
t1 = time()
print(f"BRF tuning time: {(t1-t0)/60:.2f} minutes")
print("Best BRF params:", rs_brf.best_params_)
print("Best BRF CV ROC-AUC: {:.4f}".format(rs_brf.best_score_))

best_brf = rs_brf.best_estimator_

# ---------------------------
# 2) TUNE XGBoost
# ---------------------------
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1)

xgb_param_dist = {
    'n_estimators': [200, 400, 600],
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.5],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2]
}

print("\nTuning XGBoost (this may take several minutes)...")
rs_xgb = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=xgb_param_dist,
    n_iter=20,               # increase for better search
    scoring='roc_auc',
    cv=cv,
    verbose=1,
    random_state=42,
    n_jobs=-1,
    return_train_score=False
)

# set scale_pos_weight via constructor param in fit call by using set_params before fitting
rs_xgb.estimator.set_params(scale_pos_weight=scale_pos_weight)

t0 = time()
rs_xgb.fit(X_train, y_train)
t1 = time()
print(f"XGB tuning time: {(t1-t0)/60:.2f} minutes")
print("Best XGB params:", rs_xgb.best_params_)
print("Best XGB CV ROC-AUC: {:.4f}".format(rs_xgb.best_score_))

best_xgb = rs_xgb.best_estimator_

# ---------------------------
# 3) Build Hybrid Ensemble (Stacking BRF + XGB)
# ---------------------------
print("\nBuilding stacked ensemble with best BRF and best XGB...")

stack = StackingClassifier(
    estimators=[('brf', best_brf), ('xgb', best_xgb)],
    final_estimator=LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
    cv=cv,
    n_jobs=-1,
    passthrough=False
)

t0 = time()
stack.fit(X_train, y_train)
t1 = time()
print("Stack fit time: {:.2f}s".format(t1-t0))

# ---------------------------
# 4) Cross-validated performance (on training folds) - ROC AUC
# ---------------------------
def cv_roc_auc(estimator, X, y, cv):
    scores = cross_val_score(estimator, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
    return scores.mean(), scores.std()

print("\nCross-validated ROC-AUC ({}-fold):".format(cv_folds))
for name, est in [('Best BRF', best_brf), ('Best XGB', best_xgb), ('Stacked Ensemble', stack)]:
    mean_auc, std_auc = cv_roc_auc(est, X_train, y_train, cv)
    print(f"{name}: mean ROC-AUC = {mean_auc:.4f} (Â± {std_auc:.4f})")

# ---------------------------
# 5) Evaluate on HOLD-OUT TEST SET
# ---------------------------
def evaluate_on_test(estimator, X_test, y_test, name):
    y_pred = estimator.predict(X_test)
    # some estimators (stack) may not implement predict_proba for binary -> check
    try:
        y_prob = estimator.predict_proba(X_test)[:,1]
    except Exception:
        # fallback: decision_function -> scaled to [0,1] via minmax
        try:
            scores = estimator.decision_function(X_test)
            # scale
            y_prob = (scores - scores.min()) / (scores.max() - scores.min() + 1e-9)
        except Exception:
            y_prob = np.zeros(len(y_pred))
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob) if y_prob.sum() > 0 else np.nan
    print(f"\n--- {name} ---")
    print("Accuracy :", acc)
    print("Precision:", prec)
    print("Recall   :", rec)
    print("F1-score :", f1)
    print("ROC-AUC  :", auc)
    print("\nClassification report:\n", classification_report(y_test, y_pred, digits=4))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title(f'Confusion Matrix: {name}')
    plt.show()
    return {'Model': name, 'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1': f1, 'ROC_AUC': auc}

results = []
results.append(evaluate_on_test(best_brf, X_test, y_test, 'Best BalancedRandomForest'))
results.append(evaluate_on_test(best_xgb, X_test, y_test, 'Best XGBoost'))
results.append(evaluate_on_test(stack, X_test, y_test, 'Stacked Ensemble (BRF + XGB)'))

# ---------------------------
# 6) Feature importances (BRF and XGB)
# ---------------------------
print("\nTop feature importances (Best BRF):")
try:
    imp_brf = pd.Series(best_brf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
    display(imp_brf.head(20))
    plt.figure(figsize=(8,6))
    sns.barplot(x=imp_brf.head(15).values, y=imp_brf.head(15).index)
    plt.title("Best BRF - Top 15 Importances")
    plt.show()
except Exception as e:
    print("BRF importances not available:", e)

print("\nTop feature importances (Best XGBoost):")
try:
    imp_xgb = pd.Series(best_xgb.feature_importances_, index=X_train.columns).sort_values(ascending=False)
    display(imp_xgb.head(20))
    plt.figure(figsize=(8,6))
    sns.barplot(x=imp_xgb.head(15).values, y=imp_xgb.head(15).index)
    plt.title("Best XGBoost - Top 15 Importances")
    plt.show()
except Exception as e:
    print("XGB importances not available:", e)

# ---------------------------
# 7) Summary table
# ---------------------------
summary_df = pd.DataFrame(results).set_index('Model')
print("\nSummary on hold-out test set:")
display(summary_df.round(4))
```

```python
# ============================
# Hypothesis testing & validation notebook
# ============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import ttest_ind, chi2_contingency, spearmanr, levene
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Helper: pretty print
pd.set_option('display.float_format', lambda x: f"{x:.4f}")

# Ensure df exists
assert 'Churn' in df.columns, "df must contain 'Churn'"

# Short aliases
data = df.copy()
# if Churn is numeric 0/1 ensure int
data['Churn'] = data['Churn'].astype(int)

# Quick group helpers
churn = data[data['Churn'] == 1]
nonchurn = data[data['Churn'] == 0]

# Safe column presence check (adjust names if different)
needed_cols = [
    'MonthlyCharges','TotalCharges','tenure','ContractDuration','Contract','AutoPayment',
    'TechEngagement','PriceVsDuration','LoyaltyScore','PaperlessBilling',
    'InternetService_Fiber optic','SeniorCitizen','Partner','Dependents'
]
for c in needed_cols:
    if c not in data.columns:
        print(f"WARNING: column '{c}' not found in df. Please rename accordingly.")

# ---------------------------
# Utility functions
# ---------------------------
def chi2_test_cat(col):
    ct = pd.crosstab(data[col], data['Churn'])
    chi2, p, dof, ex = chi2_contingency(ct)
    return chi2, p, ct

def ttest_numeric(col):
    x1 = churn[col].dropna()
    x0 = nonchurn[col].dropna()
    stat, p = ttest_ind(x1, x0, equal_var=False)
    return stat, p, x1.mean(), x0.mean(), len(x1), len(x0)

def spearman_test(col):
    rho, p = spearmanr(data[col], data['Churn'])
    return rho, p

# ---------------------------
# 1) "Customers with higher monthly charges are more likely to churn."
#    -> t-test & Spearman correlation
# ---------------------------
print("\n1) MonthlyCharges vs Churn")
stat, p, mean_churn, mean_nonchurn, n1, n0 = ttest_numeric('MonthlyCharges')
rho, rho_p = spearman_test('MonthlyCharges')
print(f"t-statistic={stat:.4f}, p-value={p:.4e}")
print(f"Mean MonthlyCharges (churn)={mean_churn:.4f}, (non-churn)={mean_nonchurn:.4f} | n_churn={n1}, n_nonchurn={n0}")
print(f"Spearman rho={rho:.4f}, p={rho_p:.4e}")
# plot
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=data)
plt.title('MonthlyCharges by Churn')
plt.show()

# ---------------------------
# 2) "Month-to-month contracts lead to higher churn."
#    -> chi-square on Contract (if encoded numeric, interpret first)
# ---------------------------
print("\n2) Contract type vs Churn (Chi-square)")
# If Contract is categorical text, use it; if numeric (encoded) show mapping count table
if data['Contract'].dtype == 'object':
    chi2, p, ct = chi2_test_cat('Contract')
    print(f"Chi2={chi2:.4f}, p={p:.4e}")
    print("Crosstab:")
    display(ct)
else:
    # if Contract numeric assume encoding: 0/1/2 or similar
    chi2, p, ct = chi2_test_cat('Contract')
    print(f"Chi2={chi2:.4f}, p={p:.4e}")
    display(ct)
sns.barplot(x='Contract', y='Churn', data=data.astype({'Contract':'category'}).groupby('Contract', as_index=False)['Churn'].mean())
plt.title('Churn rate by Contract')
plt.show()

# ---------------------------
# 3) "Senior citizens churn more often."
#    -> chi-square or t-test on SeniorCitizen
# ---------------------------
print("\n3) SeniorCitizen vs Churn")
chi2, p, ct = chi2_test_cat('SeniorCitizen')
print(f"Chi2={chi2:.4f}, p={p:.4e}")
display(ct)
print("Churn rate senior:", data.groupby('SeniorCitizen')['Churn'].mean().to_dict())

# ---------------------------
# 4) "Auto-paying customers churn less than those using manual payment methods (Electronic check)."
#    -> chi-square on AutoPayment and PaymentMethod_Electronic check
# ---------------------------
print("\n4) AutoPayment vs Churn")
if 'AutoPayment' in data.columns:
    chi2, p, ct = chi2_test_cat('AutoPayment')
    print(f"Chi2={chi2:.4f}, p={p:.4e}")
    display(ct)
    print("Churn rates by AutoPayment:", data.groupby('AutoPayment')['Churn'].mean().to_dict())
else:
    print("AutoPayment column missing")

if 'PaymentMethod_Electronic check' in data.columns:
    print("\nPaymentMethod_Electronic check vs Churn")
    chi2, p, ct = chi2_test_cat('PaymentMethod_Electronic check')
    print(f"Chi2={chi2:.4f}, p={p:.4e}")
    display(ct)
    print("Churn rates by Electronic check:", data.groupby('PaymentMethod_Electronic check')['Churn'].mean().to_dict())

# ---------------------------
# 5) "Customers with high charges but low add-on usage (TechEngagement) churn faster."
#    -> interaction numeric: compare groups with high MonthlyCharges & low TechEngagement
# ---------------------------
print("\n5) High MonthlyCharges & Low TechEngagement -> churn")
# create high/low by median
data['HighCharge'] = (data['MonthlyCharges'] >= data['MonthlyCharges'].median()).astype(int)
data['LowTechEng'] = (data['TechEngagement'] <= data['TechEngagement'].median()).astype(int)
ct = pd.crosstab(index=data['HighCharge']*10 + data['LowTechEng'], columns=data['Churn'])
ct.index = ['LowCharge_HighTech','LowCharge_LowTech','HighCharge_HighTech','HighCharge_LowTech']
print(ct)
# compute churn rates by segment
seg = data.groupby(['HighCharge','LowTechEng'])['Churn'].agg(['mean','count']).reset_index()
seg['segment'] = seg['HighCharge'].astype(str) + '_' + seg['LowTechEng'].astype(str)
display(seg[['segment','mean','count']])
# test: is churn rate in HighCharge_LowTech > others? use proportions test (approx with chi2)
chi2, p, _, _ = chi2_contingency(pd.crosstab((data['HighCharge']==1)&(data['LowTechEng']==1), data['Churn']))
print("Chi2 for HighCharge&LowTech vs others:", chi2, p)

# ---------------------------
# 6) "Paperless billing + short contracts amplify churn â€” digital-savvy customers are more price-sensitive."
#    -> logistic regression with interaction: PaperlessBilling * ContractDuration (short)
# ---------------------------
print("\n6) PaperlessBilling * ContractDuration interaction in logistic regression")
# Make PaperlessBilling numeric if necessary
data['PaperlessBilling'] = data['PaperlessBilling'].astype(int)
# create short_contract flag (tenure or ContractDuration threshold; assume ContractDuration exists)
data['ShortContract'] = (data['ContractDuration'] <= data['ContractDuration'].median()).astype(int)
# fit logistic regression
formula = 'Churn ~ PaperlessBilling + ContractDuration + PaperlessBilling:ContractDuration + PriceVsDuration + TechEngagement'
model = smf.logit(formula=formula, data=data).fit(disp=False)
print(model.summary())
# interpret interaction: check p-value
print("Interaction coef p-value:", model.pvalues['PaperlessBilling:ContractDuration'])

# ---------------------------
# 7) "Partnered or dependent customers have lower churn"
#    -> chi-square on Partner and Dependents
# ---------------------------
print("\n7) Partner / Dependents vs Churn")
for col in ['Partner','Dependents']:
    if col in data.columns:
        chi2, p, ct = chi2_test_cat(col)
        print(f"{col} -> chi2={chi2:.4f}, p={p:.4e}")
        display(ct)
        print("Churn rates:", data.groupby(col)['Churn'].mean().to_dict())

# ---------------------------
# 8) "Tenure moderates the relationship between contract type and churn â€” long-tenure users on monthly contracts churn less than new monthly users."
#    -> logistic regression with interaction Tenure * Contract (or ContractDuration)
# ---------------------------
print("\n8) Tenure * Contract interaction")
# if Contract categorical, convert to category for formula
data['Contract_cat'] = data['Contract'].astype('category')
# formula with interaction: Tenure * Contract_cat
formula = 'Churn ~ tenure + C(Contract_cat) + tenure:C(Contract_cat) + LoyaltyScore'
res = smf.logit(formula=formula, data=data).fit(disp=False)
print(res.summary())
# pick interaction coefficients p-values
ints = [v for v in res.params.index if ':' in v]
print("Interaction p-values:")
print(res.pvalues[ints])

# ---------------------------
# 9) "TechEngagement moderates InternetService effect â€” fiber users with more add-ons churn less despite higher bills."
#    -> interaction TechEngagement * InternetService_Fiber optic
# ---------------------------
print("\n9) TechEngagement * InternetService_Fiber optic interaction")
if 'InternetService_Fiber optic' in data.columns:
    data['Fiber'] = data['InternetService_Fiber optic'].astype(int)
    formula = 'Churn ~ TechEngagement + Fiber + TechEngagement:Fiber + PriceVsDuration + MonthlyCharges'
    mod = smf.logit(formula=formula, data=data).fit(disp=False)
    print(mod.summary())
    print("Interaction p-value:", mod.pvalues['TechEngagement:Fiber'])
else:
    print("InternetService_Fiber optic column missing")

# ---------------------------
# 10) "Senior Citizens + Fiber + Electronic check have highest churn combination."
#     -> create group and compare churn rate
# ---------------------------
print("\n10) Senior + Fiber + ElectronicCheck combination")
comb_cols = []
if 'InternetService_Fiber optic' in data.columns:
    comb_cols.append(data['InternetService_Fiber optic'].astype(int))
else:
    print("Fiber col missing")
if 'PaymentMethod_Electronic check' in data.columns:
    comb_cols.append(data['PaymentMethod_Electronic check'].astype(int))
else:
    print("Electronic check col missing")
# SeniorCitizen exists
if len(comb_cols) >= 1:
    data['HighRiskCombo'] = ((data['SeniorCitizen']==1) & (data.get('InternetService_Fiber optic',0)==1) & (data.get('PaymentMethod_Electronic check',0)==1)).astype(int)
    ct = pd.crosstab(data['HighRiskCombo'], data['Churn'])
    display(ct)
    print("Churn rates by combo:", data.groupby('HighRiskCombo')['Churn'].mean().to_dict())

# ---------------------------
# 11) "Low TotalCharges but high MonthlyCharges => early churners"
#     -> identify customers with low TotalCharges but high MonthlyCharges and compare tenure
# ---------------------------
print("\n11) Low TotalCharges & High MonthlyCharges (potential new dissatisfied customers)")
data['LowTotal_HighMonthly'] = ((data['TotalCharges'] <= data['TotalCharges'].quantile(0.25)) & (data['MonthlyCharges'] >= data['MonthlyCharges'].quantile(0.75))).astype(int)
ct = pd.crosstab(data['LowTotal_HighMonthly'], data['Churn'])
display(ct)
print("Churn rates:", data.groupby('LowTotal_HighMonthly')['Churn'].mean().to_dict())
print("Mean tenure in this group:", data.groupby('LowTotal_HighMonthly')['tenure'].mean().to_dict())

# ---------------------------
# 12) "Offering discounted bundles to customers with low TechEngagement can increase retention"
#     -> we cannot run A/B, but we can show that low TechEngagement correlates with higher churn and that cost of bundle could be computed
# ---------------------------
print("\n12) Low TechEngagement correlation with churn")
stat, pval = spearmanr(data['TechEngagement'], data['Churn'])
print(f"Spearman TechEngagement vs Churn: rho={stat:.4f}, p={pval:.4e}")
print("Churn rate by TechEngagement quartiles:")
data['Tech_q'] = pd.qcut(data['TechEngagement'], 4, labels=False, duplicates='drop')
display(data.groupby('Tech_q')['Churn'].agg(['mean','count']))

# ---------------------------
# 13) "Price dissatisfaction and lack of engagement are top drivers â€” PriceVsDuration x TechEngagement 2D segmentation"
#     -> heatmap of churn rate across binned PriceVsDuration and TechEngagement
# ---------------------------
print("\n13) 2D segmentation: PriceVsDuration vs TechEngagement")
data['Price_bin'] = pd.qcut(data['PriceVsDuration'], 4, labels=False, duplicates='drop')
data['Tech_bin'] = pd.qcut(data['TechEngagement'], 4, labels=False, duplicates='drop')
pivot = data.pivot_table(index='Price_bin', columns='Tech_bin', values='Churn', aggfunc='mean')
print("Churn rate heatmap (Price bin rows, Tech bin cols):")
display(pivot)
plt.figure(figsize=(6,5))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='Reds')
plt.title('Churn rate across PriceVsDuration x TechEngagement bins')
plt.show()

# ---------------------------
# 14) "Paperless + Fiber + Month-to-month = high churn cluster"
#     -> test churn rate for that segment
# ---------------------------
print("\n14) Paperless + Fiber + Month-to-month cluster")
if 'InternetService_Fiber optic' in data.columns and 'PaperlessBilling' in data.columns:
    cond = (data['PaperlessBilling']==1) & (data['InternetService_Fiber optic']==1) & (data['Contract']==1)  # adjust Contract code for month-to-month if needed
    # if Contract not encoded like that, user should map contract values
    print("Segment size:", cond.sum())
    print("Churn rate in this segment:", data.loc[cond,'Churn'].mean())
    print("Overall churn rate:", data['Churn'].mean())
else:
    print("Columns for this test missing or Contract encoding unclear")

# ---------------------------
# 15) "Churners have higher variance in MonthlyCharges than non-churners"
#     -> Levene test for variance
# ---------------------------
print("\n15) Variance in MonthlyCharges (Levene test)")
stat, p = levene(churn['MonthlyCharges'].dropna(), nonchurn['MonthlyCharges'].dropna())
print(f"Levene stat={stat:.4f}, p={p:.4e}")
print("Variance churn:", churn['MonthlyCharges'].var(), "variance nonchurn:", nonchurn['MonthlyCharges'].var())

# ---------------------------
# 16) "Churn is interplay: create logistic model combining key features and interactions"
#     -> full logistic regression with interactions and report coefficients and p-values
# ---------------------------
print("\n16) Multivariate logistic regression with key features & interactions")
formula = ('Churn ~ PriceVsDuration + TechEngagement + LoyaltyScore + ContractDuration + AutoPayment + '
           'PriceVsDuration:TechEngagement + PriceVsDuration:ContractDuration + LoyaltyScore:AutoPayment')
model_full = smf.logit(formula=formula, data=data).fit(disp=False)
print(model_full.summary())

# ---------------------------
# 17) "Customers with short-term contracts, high PriceVsDuration, low TechEngagement form high-risk segment"
#     -> compute churn rate and support
# ---------------------------
print("\n17) Define high-risk segment and compute churn rate")
cond_hr = (data['ContractDuration'] <= data['ContractDuration'].quantile(0.25)) & \
          (data['PriceVsDuration'] >= data['PriceVsDuration'].quantile(0.75)) & \
          (data['TechEngagement'] <= data['TechEngagement'].quantile(0.25))
print("High-risk segment size:", cond_hr.sum(), " / total:", len(data))
print("High-risk churn rate:", data.loc[cond_hr,'Churn'].mean())
print("Overall churn rate:", data['Churn'].mean())

# ---------------------------
# 18) "LoyaltyScore and AutoPayment act as stabilizers"
#     -> check logistic partial effect and group churn rates
# ---------------------------
print("\n18) LoyaltyScore & AutoPayment stabilization check")
# churn rates by loyalty & autopayment
data['HighLoyal'] = (data['LoyaltyScore'] >= data['LoyaltyScore'].median()).astype(int)
display(data.groupby(['HighLoyal','AutoPayment'])['Churn'].agg(['mean','count']).reset_index())

# fit a model to measure interaction effect
formula = 'Churn ~ LoyaltyScore + AutoPayment + LoyaltyScore:AutoPayment + PriceVsDuration + TechEngagement'
stab = smf.logit(formula=formula, data=data).fit(disp=False)
print(stab.summary())

# ---------------------------

```

```python
# ---------------------------
# Fixed: Build final hypothesis summary table with correct chi2_contingency unpacking
# ---------------------------
print("\n20) Hypotheses summary (test used, p-value / metric, result)")
hypotheses = []

# 1) MonthlyCharges (use previously computed p and rho variables if available; otherwise recompute)
stat_mc, p_mc, mean1_mc, mean0_mc, n1_mc, n0_mc = ttest_numeric('MonthlyCharges')
rho_mc, rho_p_mc = spearman_test('MonthlyCharges')
hypotheses.append(('MonthlyCharges higher -> churn', 't-test & Spearman', p_mc, rho_mc, 'Supported' if (p_mc < 0.05 and rho_mc > 0) else 'Check'))

# 2) Contract (chi2)
chi2_c, p_c, ct_c = chi2_test_cat('Contract')
hypotheses.append(('Month-to-month -> higher churn', 'chi2', p_c, chi2_c, 'Supported' if p_c < 0.05 else 'No'))

# 3) SeniorCitizen (chi2)
chi2_s, p_s, ct_s = chi2_test_cat('SeniorCitizen')
hypotheses.append(('SeniorCitizen churn more', 'chi2', p_s, chi2_s, 'Supported' if p_s < 0.05 else 'No'))

# 4) AutoPayment (chi2) - only if column exists
if 'AutoPayment' in data.columns:
    chi2_ap, p_ap, ct_ap = chi2_test_cat('AutoPayment')
    hypotheses.append(('AutoPayment -> less churn', 'chi2', p_ap, chi2_ap, 'Supported' if p_ap < 0.05 else 'No'))
else:
    hypotheses.append(('AutoPayment -> less churn', 'chi2', np.nan, np.nan, 'Column missing'))

# 5) HighCharge + LowTech (chi2) â€” CORRECT UNPACKING
flag = ((data['HighCharge'] == 1) & (data['LowTechEng'] == 1))
ct_hl = pd.crosstab(flag, data['Churn'])
chi2_hl, p_hl, dof_hl, exp_hl = chi2_contingency(ct_hl)
hypotheses.append(('High charge + low tech -> churn', 'chi2', p_hl, chi2_hl, 'Supported' if p_hl < 0.05 else 'No'))

# 6) Paperless * ContractDuration interaction (logit) â€” safe retrieval
pv_pc = model.pvalues.get('PaperlessBilling:ContractDuration', np.nan) if 'model' in globals() else np.nan
coef_pc = model.params.get('PaperlessBilling:ContractDuration', np.nan) if 'model' in globals() else np.nan
hypotheses.append(('Paperless * ContractDuration interaction', 'logit interaction', pv_pc, coef_pc, 'Supported' if (not np.isnan(pv_pc) and pv_pc < 0.05) else 'Check'))

# 7) Partner / Dependents (chi2)
chi2_p, p_p, ct_p = chi2_test_cat('Partner')
chi2_d, p_d, ct_d = chi2_test_cat('Dependents')
hypotheses.append(('Partner -> lower churn', 'chi2', p_p, chi2_p, 'Supported' if p_p < 0.05 else 'No'))
hypotheses.append(('Dependents -> lower churn', 'chi2', p_d, chi2_d, 'Supported' if p_d < 0.05 else 'No'))

# 8) Tenure * Contract interaction (logit) â€” take smallest interaction p-value if present
ints = [i for i in res.pvalues.index if ':' in i] if 'res' in globals() else []
min_int_p = res.pvalues[ints].min() if (len(ints) > 0) else np.nan
hypotheses.append(('Tenure * Contract interaction', 'logit interaction', min_int_p, None, 'Supported' if (not np.isnan(min_int_p) and min_int_p < 0.05) else 'Check'))

# 9) TechEng * Fiber interaction (logit)
if 'mod' in globals() and ('TechEngagement:Fiber' in mod.pvalues.index):
    pv_tf = mod.pvalues['TechEngagement:Fiber']
    coef_tf = mod.params['TechEngagement:Fiber']
    hypotheses.append(('TechEngagement * Fiber interaction', 'logit interaction', pv_tf, coef_tf, 'Supported' if pv_tf < 0.05 else 'Check'))
else:
    hypotheses.append(('TechEngagement * Fiber interaction', 'logit interaction', np.nan, None, 'Model/interaction missing'))

# 10) Senior + Fiber + Electronic check combo (chi2)
if 'HighRiskCombo' in data.columns:
    ct_combo = pd.crosstab(data['HighRiskCombo'], data['Churn'])
    chi2_combo, p_combo, dof_combo, exp_combo = chi2_contingency(ct_combo)
    hypotheses.append(('Senior+Fiber+ElectronicCheck -> high churn', 'chi2', p_combo, chi2_combo, 'Supported' if p_combo < 0.05 else 'Check'))
else:
    hypotheses.append(('Senior+Fiber+ElectronicCheck -> high churn', 'chi2', np.nan, None, 'Column missing'))

# 11) Low Total & High Monthly (t-test on TotalCharges as proxy)
stat_lt, p_lt, mean1_lt, mean0_lt, n1_lt, n0_lt = ttest_numeric('TotalCharges')
hypotheses.append(('Low TotalCharges & High Monthly -> early churn', 't-test', p_lt, stat_lt, 'Supported' if p_lt < 0.05 else 'Check'))

# 12) Low TechEng correlation (spearman)
rho_te, p_te = spearman_test('TechEngagement')
hypotheses.append(('Low TechEng -> churn', 'spearman', p_te, rho_te, 'Supported' if p_te < 0.05 else 'No'))

# 13) PriceVsDuration x TechEng segmentation (visual)
hypotheses.append(('PriceVsDuration & TechEngagement 2D segmentation', 'heatmap', np.nan, None, 'Visual check'))

# 14) Paperless+Fiber+Month-to-month cluster (segment)
hypotheses.append(('Paperless+Fiber+Month-to-month cluster', 'segment churn rate', np.nan, None, 'Check visually'))

# 15) MonthlyCharges variance (Levene)
lev_stat, lev_p = levene(churn['MonthlyCharges'].dropna(), nonchurn['MonthlyCharges'].dropna())
hypotheses.append(('MonthlyCharges variance higher in churn', 'levene', lev_p, lev_stat, 'Supported' if lev_p < 0.05 else 'No'))

# 16) Multivariate interplay (logit) â€” use model_full if present
if 'model_full' in globals():
    min_p_full = model_full.pvalues.min()
    hypotheses.append(('Multivariate interplay (PriceVsDuration, TechEngagement, LoyaltyScore)', 'logit', min_p_full, None, 'Supported' if min_p_full < 0.05 else 'Check'))
else:
    hypotheses.append(('Multivariate interplay (PriceVsDuration, TechEngagement, LoyaltyScore)', 'logit', np.nan, None, 'Model missing'))

# Build and display summary dataframe
summary_df = pd.DataFrame(hypotheses, columns=['Hypothesis','Test','p-value','stat/coef','Conclusion'])
display(summary_df.round(4))

print("\nDone. Interpret: p-value < 0.05 => statistical support under the chosen test.")
```

```python
# Advanced visualizations for churn dataset
# Colab-ready. Make sure `df` (your DataFrame) is already defined in the environment.

# -------------------------
# Install libs (Colab)
# -------------------------
!pip install --quiet pyvis umap-learn plotly networkx

# -------------------------
# Imports
# -------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import plotly.graph_objects as go
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import umap

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,6)

# -------------------------
# Prepare readable labels
# -------------------------
# 1) PaymentMethod label from one-hot columns (if present)
pm_cols = [c for c in df.columns if c.startswith('PaymentMethod_')]
if pm_cols:
    def get_payment_label(row):
        for c in pm_cols:
            if row[c] == 1:
                return c.replace('PaymentMethod_','')
        return 'Other'
    df['PaymentMethod_label'] = df.apply(get_payment_label, axis=1)
else:
    # fallback: if there's a PaymentMethod column (text), rename to label
    if 'PaymentMethod' in df.columns:
        df['PaymentMethod_label'] = df['PaymentMethod'].astype(str)
    else:
        print("No PaymentMethod one-hot or PaymentMethod column found -> PaymentMethod_label will be 'Unknown'")
        df['PaymentMethod_label'] = 'Unknown'

# 2) Contract label mapping (adjust if your encoding differs)
# Common mapping used earlier: 0 -> Month-to-month, 1 -> One year, 2 -> Two year
CONTRACT_MAP = {0: 'Month-to-month', 1: 'One year', 2: 'Two year'}
if df['Contract'].dtype.kind in 'iu':  # integer-like
    # map and keep fallback to str form if mapping not present
    df['Contract_label'] = df['Contract'].map(CONTRACT_MAP).fillna(df['Contract'].astype(str))
else:
    df['Contract_label'] = df['Contract'].astype(str)

# 3) Fiber label from your one-hot columns or existing column
if 'InternetService_Fiber optic' in df.columns:
    df['Fiber_flag'] = df['InternetService_Fiber optic'].astype(int)
else:
    df['Fiber_flag'] = (df.get('InternetService', '').astype(str).str.contains('Fiber', na=False)).astype(int)

# -------------------------
# 1) Arc diagram: Contract <-> PaymentMethod among churners
# -------------------------
def draw_arc_diagram(df_in, cat1='Contract_label', cat2='PaymentMethod_label', target='Churn', top_k=30):
    df_churn = df_in[df_in[target] == 1]
    pairs = df_churn.groupby([cat1, cat2]).size().reset_index(name='count').sort_values('count', ascending=False)
    left_nodes = sorted(df_in[cat1].unique())
    right_nodes = sorted(df_in[cat2].unique())
    left_x = {n: i for i,n in enumerate(left_nodes)}
    right_x = {n: i + len(left_nodes) + 1 for i,n in enumerate(right_nodes)}
    y0 = 0
    fig, ax = plt.subplots(figsize=(14,4))
    # labels
    for n,x in left_x.items():
        ax.text(x, y0, n, ha='center', va='center', fontsize=10, bbox=dict(boxstyle='round', facecolor='#a6cee3', alpha=0.9))
    for n,x in right_x.items():
        ax.text(x, y0, n, ha='center', va='center', fontsize=10, bbox=dict(boxstyle='round', facecolor='#fb9a99', alpha=0.9))
    # arcs
    top = pairs.head(top_k)
    max_count = top['count'].max() if not top.empty else 1
    for _, row in top.iterrows():
        a = left_x[row[cat1]]; b = right_x[row[cat2]]
        mid = (a + b) / 2
        width = abs(b - a)
        height = 1.2 * (row['count'] / max_count) * 3
        arc = patches.Arc((mid, y0), width=width, height=height, angle=0, theta1=0, theta2=180,
                          color='black', alpha=0.25 + 0.75*(row['count']/max_count), linewidth=1.2 + 3*(row['count']/max_count))
        ax.add_patch(arc)
    ax.set_ylim(-1, 4)
    ax.axis('off')
    plt.title(f'Arc diagram (top {top_k}) â€” {cat1} â†’ {cat2} among churners')
    plt.show()

draw_arc_diagram(df)

# -------------------------
# 2) Network graph (category co-occurrence) -> interactive HTML via PyVis
# -------------------------
def build_and_show_network(df_in, cats=['Contract_label','PaymentMethod_label','InternetService_Fiber optic'], min_edge=80):
    # build co-occurrence counts between categorical values (pairwise across selected categories)
    edges = {}
    for i in range(len(cats)):
        for j in range(i+1, len(cats)):
            a = cats[i]; b = cats[j]
            # if second column is numeric flag, convert to label
            if df_in[b].dtype.kind in 'biufc' and df_in[b].nunique()<=5:
                df_in[b+'_lbl'] = df_in[b].astype(str)
                bcol = b+'_lbl'
            else:
                bcol = b
            grp = df_in.groupby([a,bcol]).size().reset_index(name='count')
            for _, r in grp.iterrows():
                u = f"{a}:{r[a]}" if ':' not in str(r[a]) else str(r[a])
                v = f"{b}:{r[bcol]}" if ':' not in str(r[bcol]) else str(r[bcol])
                edges[(u,v)] = edges.get((u,v),0) + int(r['count'])
    # build graph
    G = nx.Graph()
    for (u,v), w in edges.items():
        if w < min_edge:
            continue
        G.add_node(u, label=u)
        G.add_node(v, label=v)
        G.add_edge(u, v, weight=w)
    net = Network(height='700px', width='100%', notebook=True, bgcolor='#ffffff', font_color='black')
    net.from_nx(G)
    net.repulsion(node_distance=200, central_gravity=0.01, spring_length=200)
    out_file = 'category_network.html'
    net.show(out_file)
    print("Interactive network saved to", out_file)

# choose the categories you want to include in the network
build_and_show_network(df, cats=['Contract_label','PaymentMethod_label','InternetService_Fiber optic'], min_edge=60)

# -------------------------
# 3) UMAP stardust cluster scatter + cluster centroids & stats
# -------------------------
def umap_stardust(df_in, features=None, n_neighbors=30, min_dist=0.1, n_clusters=4):
    if features is None:
        features = ['MonthlyCharges','TotalCharges','tenure','PriceVsDuration','TechEngagement','LoyaltyScore','ContractDuration']
    X = df_in[features].fillna(0)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)
    reducer = umap.UMAP(random_state=42, n_neighbors=n_neighbors, min_dist=min_dist)
    emb = reducer.fit_transform(Xs)
    emb_df = pd.DataFrame(emb, columns=['UMAP1','UMAP2'], index=df_in.index)
    emb_df['Churn'] = df_in['Churn'].values
    emb_df['LoyaltyScore'] = df_in['LoyaltyScore'].values
    # clustering for segments
    km = KMeans(n_clusters=n_clusters, random_state=42).fit(Xs)
    emb_df['cluster'] = km.labels_
    # plot (stardust effect)
    fig = px.scatter(emb_df, x='UMAP1', y='UMAP2', color='cluster',
                     symbol='Churn', size='LoyaltyScore', opacity=0.35,
                     title='UMAP embedding (stardust) â€” clusters + churn highlight',
                     hover_data=['cluster','Churn'])
    fig.update_traces(marker=dict(line=dict(width=0.3, color='DarkSlateGrey')))
    fig.show()
    # cluster centroids stats
    print("\nCluster summary (centroid features & churn rate):")
    merged = pd.concat([emb_df, df_in[features]], axis=1)
    for c in sorted(merged['cluster'].unique()):
        sub = merged[merged['cluster']==c]
        churn_rate = sub['Churn'].mean()
        centroid = sub[features].mean().to_dict()
        print(f"Cluster {c}: size={len(sub)}, churn_rate={churn_rate:.3f}")
        print(" centroid (sample):", {k: round(v,3) for k,v in list(centroid.items())[:5]})
        print('-'*60)
    return emb_df

emb_df = umap_stardust(df, n_clusters=4)

# -------------------------
# 4) Sankey: Contract -> PaymentMethod -> Churn (Plotly)
# -------------------------
def sankey_contract_payment_churn(df_in):
    contracts = list(df_in['Contract_label'].astype(str).unique())
    payments = list(df_in['PaymentMethod_label'].astype(str).unique())
    churn_labels = ['Stay','Churn']
    nodes = contracts + payments + churn_labels
    node_idx = {n:i for i,n in enumerate(nodes)}
    # edges contract -> payment
    c_to_p = df_in.groupby(['Contract_label','PaymentMethod_label']).size().reset_index(name='count')
    sources=[]; targets=[]; values=[]
    for _, r in c_to_p.iterrows():
        sources.append(node_idx[r['Contract_label']])
        targets.append(node_idx[r['PaymentMethod_label']])
        values.append(int(r['count']))
    # edges payment -> churn
    p_to_c = df_in.groupby(['PaymentMethod_label','Churn']).size().reset_index(name='count')
    for _, r in p_to_c.iterrows():
        sources.append(node_idx[r['PaymentMethod_label']])
        targets.append(node_idx['Churn']==1 and node_idx['Churn'] if False else None)  # placeholder to compute
    # Build mapping for churn targets
    # we'll map 0->Stay,1->Churn accordingly
    # Rebuild p_to_c loop properly:
    for _, r in p_to_c.iterrows():
        sources.append(node_idx[r['PaymentMethod_label']])
        tgt = node_idx['Churn'] if False else node_idx['Churn'] if False else None
    # easier approach: build mapping directly
    # fix: rebuild nodes so churn labels last two entries are known
    nodes = contracts + payments + churn_labels
    node_idx = {n:i for i,n in enumerate(nodes)}
    sources=[]; targets=[]; values=[]
    for _, r in c_to_p.iterrows():
        sources.append(node_idx[r['Contract_label']])
        targets.append(node_idx[r['PaymentMethod_label']])
        values.append(int(r['count']))
    p_to_c = df_in.groupby(['PaymentMethod_label','Churn']).size().reset_index(name='count')
    for _, r in p_to_c.iterrows():
        sources.append(node_idx[r['PaymentMethod_label']])
        tgt_label = 'Churn' if int(r['Churn'])==1 else 'Stay'
        targets.append(node_idx[tgt_label])
        values.append(int(r['count']))
    # plot sankey
    fig = go.Figure(data=[go.Sankey(
        node=dict(pad=15, thickness=20, line=dict(color="black", width=0.5), label=nodes),
        link=dict(source=sources, target=targets, value=values)
    )])
    fig.update_layout(title_text="Sankey: Contract -> PaymentMethod -> Churn/Stay", font_size=12)
    fig.show()

sankey_contract_payment_churn(df)

# -------------------------
# Save snapshot PNGs for PPT (optional)
# -------------------------
# You can save static screenshots of Matplotlib plots (arc) or export Plotly figs as image (requires kaleido)
# Example for arc diagram already plotted; to save:
# fig.savefig('arc_diagram.png', dpi=200)
# For Plotly export if needed:
# !pip install -q kaleido
# fig.write_image('sankey.png', scale=2)

print("All visuals created. - Arc (matplotlib), Network (pyvis html), UMAP (plotly), Sankey (plotly).")
print("Open 'category_network.html' from the notebook filesystem to interact with the network.")


```
